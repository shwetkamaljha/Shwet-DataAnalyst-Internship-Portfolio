# Task 3: Deep-Dive Analysis & Interactive Dashboard

**Objective**- To perform in-depth data analysis, identify key business insights, and build an interactive dashboard for decision-making.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("task1_cleaned_superstore.csv")
df.head()
```

```python
df.isnull().sum()
df.duplicated().sum()
```

## Data Validation

The dataset used in this analysis was already cleaned in previous tasks.

However, a quick validation was performed to ensure:
- No missing values are present
- No duplicate records exist

The dataset is clean and ready for analysis.

**KPI Section**

```python
total_revenue = df['sales'].sum()
total_profit = df['profit'].sum()
total_orders = df.shape[0]
average_order_value = df['sales'].mean()

print("Total Revenue:", total_revenue)
print("Total Profit:", total_profit)
print("Total Orders:", total_orders)
print("Average Order Value:", average_order_value)
```

## Key Performance Indicators (KPIs)

- Total Revenue: Overall sales generated from all transactions
- Total Profit: Total profit earned by the business
- Total Orders: Total number of transactions/orders
- Average Order Value: Average revenue per order

# **Deep Dive Analysis**

**Monthly Sales Trend**

```python
df['order_date'] = pd.to_datetime(df['order_date'])

monthly_sales = df.groupby(df['order_date'].dt.month)['sales'].sum()

monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

## Monthly Sales Analysis

This analysis shows how sales vary across different months and helps identify seasonal trends.

**Category Analysis**

```python
category_sales = df.groupby('category')['sales'].sum()

category_sales.plot(kind='bar')
plt.title("Sales by Category")
plt.show()
```

## Category-wise Analysis

This analysis identifies which category contributes the most to total revenue.

**Top Products**

```python
top_products = df.groupby('product_name')['sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar')
plt.title("Top 10 Products")
plt.show()
```

## Top Products Analysis

Top-performing products are identified based on highest sales contribution.

## Observations

1. The total revenue generated is ₹2,297,200, indicating strong overall business performance.

2. The total profit is ₹286,400, showing that the business is operating profitably.

3. Sales show a peak in the month of November, indicating high customer demand during this period, possibly due to seasonal factors or festive sales.

4. The category "Technology" contributes the highest revenue among all categories, making it the most significant segment for the business.

5. The top-selling product is "Canon ImageCLASS 2200 Advanced Copier", generating the highest sales among all products.

6. The sales trend shows a gradual increase over time, indicating consistent business growth.
